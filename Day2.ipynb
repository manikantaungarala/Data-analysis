```python
#there are 5 subjects psychology nanotechnology medicine on-cology antho-cology two students have applied to do masters in any of these five fields with a line plot show in which field they can excel the cutoff  for each subject is 70
stu1=[67,45,73,71,57]
stu2=[43,67,68,53,85]
sub=["psychology","nanotecnology","medicine","oncology","anthocology"]
cut=[70,70,70,70,70]
#plt.plottype(x,y,label,color,marker,linewidth,linestyle)
plt.scatter(sub,stu1,label="student1",color="k",marker="p") #scatter means dot
plt.scatter(sub,stu2,label="student2",color="b",marker="H")#plot means lines
plt.plot(cut,label="cutoff",color="r",linestyle=":",linewidth=3)
plt.legend()
plt.xlabel("subjects")
plt.ylabel("marks")

```

```python
#INDvsAUS match is live india gives target 298 as aus,
#the runs for every 5overs is given as follows ind={25,51,84,131,160,189,220,250,267,297}
#australia completed 25 overs aus={15,41,94,110,151}
#plot overs vs runs for both countries with ind blue line and aus yellow line
#with required labels and legends the marker for india is flag and marker for australia is plus,take overs from arange function create runs as arrays
#linewidth=2
from matplotlib import pyplot as plt
import numpy as np
o=np.arange(5,51,5)
ind=np.array([25,51,84,131,160,189,220,250,267,297])
aus=np.array([15,41,94,110,151])
o2=np.arange(5,26,5)
plt.plot(o,ind,label="ind",color="b",marker=">",linewidth=2)
plt.plot(o2,aus,label="aus",color="y",marker="+",linewidth=2)
plt.title("INDvsAUS RUNS ANALYSIS")
plt.legend()
plt.xlabel("overs")
plt.ylabel("runs")
plt.grid()
```
